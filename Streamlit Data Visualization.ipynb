```python
def Outlier1(x):
  out=[]
  Q1=4.000000
  Q3=4.500000
  IQR=Q3-Q1
  upper=Q3+(1.5*IQR)
  lower=Q1-(1.5*IQR)
  for i in x:
    if (i>upper) or (i<lower):
      out.append(i)
  return out
```

```python
Outlier1(Data['Rating'])
```
